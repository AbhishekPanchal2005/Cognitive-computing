```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Part I: Randomized Sales Data Generation (NumPy)
roll_number = 102317167  
np.random.seed(roll_number)

# Generate Sales Data
sales_data = np.random.randint(1000, 5001, size=(12, 4))
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Convert to DataFrame
sales_df = pd.DataFrame(sales_data, columns=categories, index=months)
```

```python

# Part II: Data Manipulation and Analysis (Pandas)

# Display first 5 rows and summary statistics
print(sales_df.head())
print(sales_df.describe())

# Total sales per category and total sales per month
total_sales_category = sales_df.sum()
total_sales_month = sales_df.sum(axis=1)

# Average sales growth between consecutive months for each category
avg_sales_growth = sales_df.pct_change().mean()

# Add Total Sales and Growth Rate columns
sales_df['Total Sales'] = total_sales_month
sales_df['Growth Rate'] = sales_df['Total Sales'].pct_change() * 100

# Discount based on odd roll number (102317167)
sales_df['Clothing'] = sales_df['Clothing'] * 0.85
```

```python
# Part III: Visualizations (Matplotlib and Seaborn)
# Monthly sales trends for each category
plt.figure(figsize=(12, 6))
for category in categories:
    plt.plot(sales_df.index, sales_df[category], label=category)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Trends')
plt.legend()
plt.show()

# Box plots for sales distribution of each category
plt.figure(figsize=(12, 6))
sns.boxplot(data=sales_df[categories])
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales Distribution by Category')
plt.show()
```

```python
# Q2
array = np.array([[1, -2, 3], [-4, 5, -6]])

# Element-wise absolute value
abs_array = np.abs(array)

# Percentiles
percentiles = {
    "25th": np.percentile(array, 25, axis=None),
    "50th": np.percentile(array, 50, axis=None),
    "75th": np.percentile(array, 75, axis=None),
}

# Mean, Median, and Standard Deviation
mean = np.mean(array, axis=None)
median = np.median(array, axis=None)
std_dev = np.std(array, axis=None)

print("Absolute Value:\n", abs_array)
print("Percentiles:\n", percentiles)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
```

```python
#Q3
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

# Floor, Ceiling, and Truncated values
floor_values = np.floor(a)
ceiling_values = np.ceil(a)
truncated_values = np.trunc(a)
rounded_values = np.round(a)

print("Floor values:", floor_values)
print("Ceiling values:", ceiling_values)
print("Truncated values:", truncated_values)
print("Rounded values:", rounded_values)
```

```python
#Q4
def swap_elements(lst, index1, index2):
    temp = lst[index1]
    lst[index1] = lst[index2]
    lst[index2] = temp
    return lst

lst = [1, 2, 3, 4]
swap_elements(lst, 1, 3)
```

```python
#Q5
def swap_elements_in_set(s, index1, index2):
    lst = list(s)
    temp = lst[index1]
    lst[index1] = lst[index2]
    lst[index2] = temp
    return set(lst)

s = {1, 2, 3, 4}
swapped_set = swap_elements_in_set(s, 1, 3)  
print(swapped_set)
```
